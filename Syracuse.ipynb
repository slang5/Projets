```python
import seaborn as sb
import numpy as np
def Process(valeur):
    mat = [valeur] 
    Vrai = True
    count = 0
    while Vrai:
        val = mat[count]
        if val == 1:
            Vrai = False
            return np.array(mat)
        elif val % 2 == 0:
            val = val/2
            mat.append(val)
        else:
            val = 3*val+1
            mat.append(val)
        count += 1
```

```python
Resultat1 = [(Process(i**2)) for i in range(100,200,2)]
sb.lineplot(Resultat1).legend([],[])
```

```python
Resultat2 = [np.log(i) for i in Resultat1]
sb.lineplot(Resultat2).legend([],[])
```
